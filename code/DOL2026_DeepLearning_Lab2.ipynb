```python
# 🔧 Install
# !pip install torch torchvision timm matplotlib seaborn scikit-learn

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms

import timm
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.metrics import accuracy_score
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)
```

```python
# 📦 Fashion MNIST

transform = transforms.Compose([
    transforms.Resize((224,224)),  # for big models
    transforms.Grayscale(num_output_channels=3),  # convert to 3 channels
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_dataset = torchvision.datasets.FashionMNIST(
    root='./data', train=True, download=True, transform=transform)

test_dataset = torchvision.datasets.FashionMNIST(
    root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=64)
```

```python
# 🔍 Preview

classes = train_dataset.classes

def show_samples():
    imgs, labels = next(iter(train_loader))
    fig, axes = plt.subplots(1, 8, figsize=(12,3))

    for i in range(8):
        axes[i].imshow(imgs[i].permute(1,2,0))
        axes[i].set_title(classes[labels[i]])
        axes[i].axis('off')

    plt.show()

# Print your sample dataset
```

```python
def train_model(model, name, epochs=2):
    optimizer = optim.Adam(model.parameters(), lr=1e-4)
    criterion = nn.CrossEntropyLoss()

    for epoch in range(epochs):
        model.train()
        total_loss = 0

        for imgs, labels in tqdm(train_loader, desc=f"{name} Epoch {epoch+1}"):
            imgs, labels = imgs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        print(f"{name} Loss: {total_loss:.4f}")

    return model
```

```python
def evaluate(model):
    model.eval()
    preds, trues = [], []

    with torch.no_grad():
        for imgs, labels in test_loader:
            imgs = imgs.to(device)
            outputs = model(imgs)
            pred = torch.argmax(outputs, dim=1).cpu()

            preds.extend(pred.numpy())
            trues.extend(labels.numpy())

    return accuracy_score(trues, preds)
```

```python
results = {}

for name, model in models.items():
    trained = train_model(model, name)
    acc = evaluate(trained)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
```

```python
names = list(results.keys())
scores = list(results.values())

plt.figure(figsize=(8,5))
sns.barplot(x=names, y=scores)

plt.title("Model Comparison")
plt.ylabel("Accuracy")
plt.show()
```

```python
# 🔮 Show predictions

best_model_name = max(results, key=results.get)
model = models[best_model_name]

imgs, labels = next(iter(test_loader))

with torch.no_grad():
    outputs = model(imgs.to(device))
    preds = torch.argmax(outputs, dim=1).cpu()

fig, axes = plt.subplots(1, 6, figsize=(12,3))

for i in range(6):
    axes[i].imshow(imgs[i].permute(1,2,0))
    axes[i].set_title(f"P:{classes[preds[i]]}\nT:{classes[labels[i]]}")
    axes[i].axis('off')

plt.show()

print("Best model:", best_model_name)
```
